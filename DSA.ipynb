```python
#1
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def reverse_linked_list(head):
    prev = None
    current = head
    
    while current is not None:
        next_node = current.next  # Save the next node
        current.next = prev       # Reverse the link
        prev = current            # Move prev to the current node
        current = next_node       # Move to the next node
    
    return prev  # New head of the reversed list

# Helper function to print the linked list
def print_list(head):
    current = head
    while current:
        print(current.value, end=' -> ')
        current = current.next
    print('None')

# Example usage:
# Creating a linked list: 1 -> 2 -> 3 -> 4 -> 5 -> None
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

print("Original list:")
print_list(head)

# Reverse the linked list
reversed_head = reverse_linked_list(head)

print("Reversed list:")
print_list(reversed_head)
```

```python
#2
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def merge_two_sorted_lists(l1, l2):
    # Create a dummy node to serve as the starting point of the merged list
    dummy = ListNode()
    current = dummy
    
    # Traverse both lists
    while l1 is not None and l2 is not None:
        if l1.value <= l2.value:
            current.next = l1
            l1 = l1.next
        else:
            current.next = l2
            l2 = l2.next
        current = current.next
    
    # Append the remaining nodes of l1 or l2 (only one of these will be non-null)
    if l1 is not None:
        current.next = l1
    if l2 is not None:
        current.next = l2
    
    # Return the merged list starting from the node after the dummy node
    return dummy.next

# Helper function to print the linked list
def print_list(head):
    current = head
    while current:
        print(current.value, end=' -> ')
        current = current.next
    print('None')

# Example usage:
# Creating two sorted linked lists
# List 1: 1 -> 3 -> 5 -> None
# List 2: 2 -> 4 -> 6 -> None

l1 = ListNode(1)
l1.next = ListNode(3)
l1.next.next = ListNode(5)

l2 = ListNode(2)
l2.next = ListNode(4)
l2.next.next = ListNode(6)

print("List 1:")
print_list(l1)

print("List 2:")
print_list(l2)

# Merge the lists
merged_list = merge_two_sorted_lists(l1, l2)

print("Merged list:")
print_list(merged_list)
```

```python
#9
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next
def is_palindrome(head):
    if not head or not head.next:
        return True

    # Step 1: Find the middle of the list
    def find_middle(node):
        slow = fast = node
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        return slow

    # Step 2: Reverse the second half of the list
    def reverse_list(node):
        prev = None
        while node:
            next_node = node.next
            node.next = prev
            prev = node
            node = next_node
        return prev

    # Step 3: Compare the first half and the reversed second half
    middle = find_middle(head)
    second_half = reverse_list(middle)
    first_half = head

    while second_half:
        if first_half.value != second_half.value:
            return False
        first_half = first_half.next
        second_half = second_half.next

    return True
def is_palindrome_stack(head):
    if not head:
        return True
    
    stack = []
    current = head
    
    # Step 1: Traverse the list and push values onto the stack
    while current:
        stack.append(current.value)
        current = current.next
    
    # Step 2: Traverse the list again and compare with the stack
    current = head
    while current:
        if current.value != stack.pop():
            return False
        current = current.next
    
    return True
# Helper function to create a linked list from a list of values
def create_linked_list(values):
    if not values:
        return None
    head = ListNode(values[0])
    current = head
    for value in values[1:]:
        current.next = ListNode(value)
        current = current.next
    return head

# Helper function to print the linked list
def print_linked_list(head):
    current = head
    while current:
        print(current.value, end=' -> ')
        current = current.next
    print('None')

# Example usage
# Creating a linked list: 1 -> 2 -> 3 -> 2 -> 1 -> None
head = create_linked_list([1, 2, 3, 2, 1])

print("List:")
print_linked_list(head)

# Checking if the list is a palindrome
print("Is palindrome (Reverse approach)?", is_palindrome(head))
print("Is palindrome (Stack approach)?", is_palindrome_stack(head))
```

```python
#11
class Node:
    def __init__(self, val=0, next=None, prev=None, child=None):
        self.val = val
        self.next = next
        self.prev = prev
        self.child = child

def flatten(head):
    def flatten_and_return_last(node):
        current = node
        last = None
        
        while current:
            next_node = current.next
            if current.child:
                child_last = flatten_and_return_last(current.child)
                
                # Connect the current node to the child list
                current.next = current.child
                current.child.prev = current
                
                # Connect the last node of the child list to the next node
                if child_last:
                    child_last.next = next_node
                    if next_node:
                        next_node.prev = child_last
                
                # Remove the child pointer
                current.child = None
            
            last = current
            current = next_node
        
        return last
    
    if not head:
        return None
    
    flatten_and_return_last(head)
    return head
# Helper function to print the flattened list
def print_list(head):
    while head:
        print(head.val, end=" ")
        head = head.next
    print()

# Example creation of a multilevel doubly linked list
head = Node(1)
head.next = Node(2)
head.next.prev = head
head.next.next = Node(3)
head.next.next.prev = head.next
head.child = Node(7)
head.child.next = Node(8)
head.child.next.prev = head.child
head.child.next.next = Node(9)
head.child.next.next.prev = head.child.next

# Flattening the list
flattened_head = flatten(head)

# Printing the flattened list
print_list(flattened_head)  # Output should be: 1 7 8 9 2 3
```

```python
#14
def search_insert_position(nums, target):
    left, right = 0, len(nums) - 1
    
    while left <= right:
        mid = (left + right) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return left
# Example usage
nums = [1, 3, 5, 6]
target = 5
print("Index of target:", search_insert_position(nums, target))  # Output: 2

target = 2
print("Index where target should be inserted:", search_insert_position(nums, target))  # Output: 1

target = 7
print("Index where target should be inserted:", search_insert_position(nums, target))  # Output: 4

target = 0
print("Index where target should be inserted:", search_insert_position(nums, target))  # Output: 0
```

```python
#15
def find_min_in_rotated_sorted_array(nums):
    left, right = 0, len(nums) - 1
    
    while left < right:
        mid = (left + right) // 2
        
        if nums[mid] > nums[right]:
            # Minimum element is in the right half
            left = mid + 1
        else:
            # Minimum element is in the left half or at mid
            right = mid
    
    # At the end of the loop, left == right and points to the minimum element
    return nums[left]
# Example usage
nums = [4, 5, 6, 7, 0, 1, 2]
print("Minimum element:", find_min_in_rotated_sorted_array(nums))  # Output: 0

nums = [11, 13, 15, 17, 19, 3, 5, 7, 9]
print("Minimum element:", find_min_in_rotated_sorted_array(nums))  # Output: 3

nums = [2, 3, 4, 5, 6, 1]
print("Minimum element:", find_min_in_rotated_sorted_array(nums))  # Output: 1

nums = [1]
print("Minimum element:", find_min_in_rotated_sorted_array(nums))  # Output: 1
```

```python
#16
def search_in_rotated_sorted_array(nums, target):
    left, right = 0, len(nums) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if nums[mid] == target:
            return mid
        
        # Determine which part is sorted
        if nums[left] <= nums[mid]:  # Left part is sorted
            if nums[left] <= target < nums[mid]:
                right = mid - 1
            else:
                left = mid + 1
        else:  # Right part is sorted
            if nums[mid] < target <= nums[right]:
                left = mid + 1
            else:
                right = mid - 1
    
    return -1
# Example usage
nums = [4, 5, 6, 7, 0, 1, 2]
target = 0
print("Index of target:", search_in_rotated_sorted_array(nums, target))  # Output: 4

target = 3
print("Index of target:", search_in_rotated_sorted_array(nums, target))  # Output: -1

nums = [11, 13, 15, 17, 19, 3, 5, 7, 9]
target = 7
print("Index of target:", search_in_rotated_sorted_array(nums, target))  # Output: 6

target = 15
print("Index of target:", search_in_rotated_sorted_array(nums, target))  # Output: 2

nums = [2, 3, 4, 5, 6, 1]
target = 1
print("Index of target:", search_in_rotated_sorted_array(nums, target))  # Output: 5
```

```python
#17
def find_peak_element(nums):
    left, right = 0, len(nums) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        # Check if mid is a peak element
        if (mid == 0 or nums[mid] > nums[mid - 1]) and (mid == len(nums) - 1 or nums[mid] > nums[mid + 1]):
            return mid
        
        # Decide the direction to move
        if mid > 0 and nums[mid] < nums[mid - 1]:
            right = mid - 1
        else:
            left = mid + 1
    
    return -1
# Example usage
nums = [1, 3, 20, 4, 1]
print("Index of peak element:", find_peak_element(nums))  # Output: 2 (Peak element is 20)

nums = [1, 2, 3, 4, 5]
print("Index of peak element:", find_peak_element(nums))  # Output: 4 (Peak element is 5)

nums = [5, 10, 20, 15]
print("Index of peak element:", find_peak_element(nums))  # Output: 2 (Peak element is 20)

nums = [1, 2, 3]
print("Index of peak element:", find_peak_element(nums))  # Output: 2 (Peak element is 3)

nums = [3, 2, 1]
print("Index of peak element:", find_peak_element(nums))  # Output: 0 (Peak element is 3)
```

```python
#18
def count_negatives(matrix):
    if not matrix or not matrix[0]:
        return 0
    
    m, n = len(matrix), len(matrix[0])
    count = 0
    row, col = 0, n - 1
    
    while row < m and col >= 0:
        if matrix[row][col] < 0:
            # All elements to the left of col in this row are negative
            count += (col + 1)
            # Move to the next row
            row += 1
        else:
            # Move left if the current element is non-negative
            col -= 1
    
    return count
# Example usage
matrix = [
    [4, 3, 2, -1],
    [3, 2, 1, -1],
    [1, 1, -1, -2],
    [-1, -1, -2, -3]
]

print("Number of negative numbers:", count_negatives(matrix))  # Output: 8

matrix = [
    [3, 2],
    [1, 0]
]

print("Number of negative numbers:", count_negatives(matrix))  # Output: 0

matrix = [
    [-1]
]

print("Number of negative numbers:", count_negatives(matrix))  # Output: 1
```

```python
#19
def search_matrix(matrix, target):
    if not matrix or not matrix[0]:
        return False

    m, n = len(matrix), len(matrix[0])
    left, right = 0, m * n - 1
    
    while left <= right:
        mid = (left + right) // 2
        mid_value = matrix[mid // n][mid % n]
        
        if mid_value == target:
            return True
        elif mid_value < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return False
# Example usage
matrix = [
    [1, 4, 7, 11],
    [2, 5, 8, 12],
    [3, 6, 9, 16],
    [10, 13, 14, 17]
]

target = 5
print("Target present:", search_matrix(matrix, target))  # Output: True

target = 20
print("Target present:", search_matrix(matrix, target))  # Output: False
```

```python
#20
def find_median_sorted_arrays(A, B):
    # Ensure A is the smaller array
    if len(A) > len(B):
        A, B = B, A

    m, n = len(A), len(B)
    imin, imax, half_len = 0, m, (m + n + 1) // 2

    while imin <= imax:
        i = (imin + imax) // 2
        j = half_len - i

        if i < m and B[j-1] > A[i]:
            imin = i + 1
        elif i > 0 and A[i-1] > B[j]:
            imax = i - 1
        else:
            # i is perfect
            if i == 0: max_of_left = B[j-1]
            elif j == 0: max_of_left = A[i-1]
            else: max_of_left = max(A[i-1], B[j-1])

            if (m + n) % 2 == 1:
                return max_of_left

            if i == m: min_of_right = B[j]
            elif j == n: min_of_right = A[i]
            else: min_of_right = min(A[i], B[j])

            return (max_of_left + min_of_right) / 2.0
# Example usage
A = [1, 3]
B = [2]

print("Median:", find_median_sorted_arrays(A, B))  # Output: 2.0

A = [1, 2]
B = [3, 4]

print("Median:", find_median_sorted_arrays(A, B))  # Output: 2.5
```

```python
#21
def next_greatest_letter(letters, target):
    left, right = 0, len(letters)
    
    while left < right:
        mid = (left + right) // 2
        if letters[mid] <= target:
            left = mid + 1
        else:
            right = mid
    
    # After binary search, left points to the smallest element greater than target
    # If left is beyond the end of the array, wrap around to the start
    return letters[left % len(letters)]
# Example usage
letters = ['c', 'f', 'j']
target = 'a'
print("Next greatest letter:", next_greatest_letter(letters, target))  # Output: 'c'

target = 'c'
print("Next greatest letter:", next_greatest_letter(letters, target))  # Output: 'f'

target = 'd'
print("Next greatest letter:", next_greatest_letter(letters, target))  # Output: 'f'

target = 'j'
print("Next greatest letter:", next_greatest_letter(letters, target))  # Output: 'c'
```

```python
#22
def sort_colors(nums):
    low, mid, high = 0, 0, len(nums) - 1
    
    while mid <= high:
        if nums[mid] == 0:
            nums[low], nums[mid] = nums[mid], nums[low]
            low += 1
            mid += 1
        elif nums[mid] == 1:
            mid += 1
        else:  # nums[mid] == 2
            nums[high], nums[mid] = nums[mid], nums[high]
            high -= 1
# Example usage
nums = [2, 0, 2, 1, 1, 0]
sort_colors(nums)
print("Sorted colors:", nums)  # Output: [0, 0, 1, 1, 2, 2]

nums = [2, 0, 1]
sort_colors(nums)
print("Sorted colors:", nums)  # Output: [0, 1, 2]
```

```python
#23
import heapq

def find_kth_largest(nums, k):
    # Create a min-heap with the first k elements
    min_heap = nums[:k]
    heapq.heapify(min_heap)
    
    for num in nums[k:]:
        if num > min_heap[0]:
            heapq.heappushpop(min_heap, num)
    
    return min_heap[0]
import random

def quickselect(nums, left, right, k_smallest):
    if left == right:
        return nums[left]

    pivot_index = random.randint(left, right)
    pivot_index = partition(nums, left, right, pivot_index)
    
    if k_smallest == pivot_index:
        return nums[k_smallest]
    elif k_smallest < pivot_index:
        return quickselect(nums, left, pivot_index - 1, k_smallest)
    else:
        return quickselect(nums, pivot_index + 1, right, k_smallest)

def partition(nums, left, right, pivot_index):
    pivot_value = nums[pivot_index]
    # Move pivot to end
    nums[pivot_index], nums[right] = nums[right], nums[pivot_index]
    store_index = left
    for i in range(left, right):
        if nums[i] < pivot_value:
            nums[store_index], nums[i] = nums[i], nums[store_index]
            store_index += 1
    # Move pivot to its final place
    nums[right], nums[store_index] = nums[store_index], nums[right]
    return store_index

def find_kth_largest(nums, k):
    size = len(nums)
    # kth largest is (size - k)th smallest
    return quickselect(nums, 0, size - 1, size - k)
def find_kth_largest(nums, k):
    nums.sort(reverse=True)
    return nums[k-1]
# Example usage
nums = [3, 2, 1, 5, 6, 4]
k = 2

print("Using heap-based approach:", find_kth_largest(nums, k))  # Output: 5
print("Using quickselect approach:", find_kth_largest(nums, k))  # Output: 5
print("Using sorting approach:", find_kth_largest(nums, k))     # Output: 5
```

```python
#28
def factorial_iterative(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
def factorial_recursive(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    if n == 0 or n == 1:
        return 1
    return n * factorial_recursive(n - 1)
import math

def factorial_builtin(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    return math.factorial(n)
# Example usage for iterative approach
print("Factorial (iterative) of 5:", factorial_iterative(5))  # Output: 120

# Example usage for recursive approach
print("Factorial (recursive) of 5:", factorial_recursive(5))  # Output: 120

# Example usage for built-in function
print("Factorial (built-in) of 5:", factorial_builtin(5))     # Output: 120
```

```python
#30
def generate_fibonacci_up_to_n(n):
    if n < 0:
        raise ValueError("Input should be a non-negative integer")
    
    fibonacci_series = []
    a, b = 0, 1
    
    while a <= n:
        fibonacci_series.append(a)
        a, b = b, a + b
    
    return fibonacci_series
# Example usage
n = 100
fibonacci_series = generate_fibonacci_up_to_n(n)
print("Fibonacci series up to", n, ":", fibonacci_series)
```

```python
#31
def power_recursive(base, exponent):
    # Base case for exponent 0
    if exponent == 0:
        return 1
    # Base case for exponent 1
    if exponent == 1:
        return base
    # Recursive case
    if exponent % 2 == 0:
        half_power = power_recursive(base, exponent // 2)
        return half_power * half_power
    else:
        return base * power_recursive(base, exponent - 1)
# Example usage
base = 2
exponent = 10
result = power_recursive(base, exponent)
print(f"{base}^{exponent} =", result)  # Output: 2^10 = 1024
```
